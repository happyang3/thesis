### Import Files

```python
import pandas as pd
from google.colab import files

# Upload files from your local file system
uploaded_form_data = files.upload()
```

```python
# Read uploaded files into pandas DataFrames
form_dfs = {}
dem_dfs = {}
for filename in uploaded_form_data.keys():
    # Assuming the file is a CSV
    form_dfs[filename] = pd.read_csv(filename)

for filename in uploaded_dem_data.keys():
    # Assuming the file is a CSV
    dem_dfs[filename] = pd.read_csv(filename)

print(form_dfs.keys())
print(dem_dfs.keys())

display(form_dfs["Hiring_company.csv"].head())
display(dem_dfs["Hiring_company_dem.csv"].head())
```

###Data Cleaning

Merging demographic data with form responses

```python
# Set the display options to show all columns
pd.set_option('display.max_columns', None)

merged_dfs = {}

# Iterate over the keys in form_dfs
for form_key in form_dfs:

    # Generate the corresponding key for dem_dfs
    dem_key = form_key.replace('.csv', '_dem.csv')

    # Check if the generated key exists in dem_dfs
    if dem_key in dem_dfs:

      dem_dfs[dem_key] = dem_dfs[dem_key].rename(columns={"Participant id": "What is your Prolific ID?"})

      # Merge the DataFrames on a common column
      merged_dfs[form_key] = pd.merge(form_dfs[form_key], dem_dfs[dem_key], on='What is your Prolific ID?', how='inner')

    else:

        print(f"No matching key found for {dem_key}")

display(merged_dfs['Hiring_company.csv'].head())
```

Drop extraneous columns

```python
for key in merged_dfs:
  merged_dfs[key] = merged_dfs[key].drop(['Timestamp', 'Summary', 'Submission id', 'Status', 'Custom study tncs accepted at', 'Started at',
                                          'Completed at', 'Reviewed at', 'Archived at'], axis=1)

display(merged_dfs['Hiring_company.csv'].head())
```

```python
# company, journalistic, interactive
changes_experience = {"I have never heard of AI.": [0,0,0], "I have heard of AI, but have never knowingly interacted with an AI system.": [0,0,0],
                      "I have used AI systems a handful of times.": [0,0,0], "I have taken at least one class on AI.": [0,0,0],
                      "I use AI systems regularly in my work or personal life.": [0,0,0], "I have designed or programmed an AI system before.": [0,0,0]}

for statement in experience_statements:
  changes_experience[statement][0] = (responses_experience[statement][2] / responses_experience[statement][3]) - (responses_experience[statement][0] / responses_experience[statement][1])
  changes_experience[statement][1] = (responses_experience[statement][6] / responses_experience[statement][7]) - (responses_experience[statement][4] / responses_experience[statement][5])
  changes_experience[statement][2] = (responses_experience[statement][10] / responses_experience[statement][11]) - (responses_experience[statement][8] / responses_experience[statement][9])

print(changes_experience)
```

```python
# Group names and number of groups
groups = ['Company', 'Journalistic', 'Interactive']
n_groups = len(groups)
n_statements = len(changes_experience)

# Prepare the figure and axes
fig, ax = plt.subplots(figsize=(10, 8), dpi=160)

# Set up the bar width and positions
bar_width = 0.15
index = np.arange(n_groups)  # the label locations
offset = bar_width * n_statements

# Colors for visual distinction
colors = plt.cm.viridis(np.linspace(0, 1, n_statements))

# Create bars for each statement
for i, (statement, values) in enumerate(changes_experience.items()):
    # Calculate the position for each group's bars
    pos = index + i * bar_width
    ax.bar(pos, values, width=bar_width, label=statement, color=colors[i])

# Customize the plot
ax.set_xlabel('Explanation Type', labelpad=10)
ax.set_ylabel('Change in Proportion of Correct Responses', labelpad=10)
#ax.set_title('Experience with AI Across Different Groups')
ax.set_xticks(index + offset / 2 - bar_width)
ax.set_xticklabels(groups)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=1)

#plt.subplots_adjust(top= 0.6)  # Adjust top margin to make room for the legend

# Layout adjustment
plt.tight_layout()

# Show plot
plt.show()
```
